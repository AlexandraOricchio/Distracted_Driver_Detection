```python
import os
import pandas as pd
import numpy as np
import operator
```

```python
dist_2016_path = os.path.join('2016 - DISTRACT.csv')
raw2016 = pd.read_csv(dist_2016_path, header = 0)

dist_2017_path = os.path.join('2017 - DISTRACT.csv')
raw2017 = pd.read_csv(dist_2017_path, header = 0)

dist_2018_path = os.path.join('2018 - DISTRACT.csv')
raw2018 = pd.read_csv(dist_2018_path, header = 0)

viol_2016_path = os.path.join('2016 - VIOLATN.csv')
raw2016_viol = pd.read_csv(viol_2016_path, header = 0)

viol_2017_path = os.path.join('2017 - VIOLATN.csv')
raw2017_viol = pd.read_csv(viol_2017_path, header = 0)

viol_2018_path = os.path.join('2018 - VIOLATN.csv')
raw2018_viol = pd.read_csv(viol_2018_path, header = 0)
```

```python
raw2016['Year'] = '2016'
raw2017['Year'] = '2017'
raw2018['Year'] = '2018'

raw2016_viol['Year'] = '2016'
raw2017_viol['Year'] = '2017'
raw2018_viol['Year'] = '2018'
```

```python
raw2018.head(5)
```

```python
big_raw = raw2016
big_raw = big_raw.append(raw2017)
big_raw = big_raw.append(raw2018)
big_raw = big_raw.reset_index()

viol_raw = raw2016_viol
viol_raw = viol_raw.append(raw2017_viol)
viol_raw = viol_raw.append(raw2018_viol)
viol_raw
```

```python
dist_legend = pd.DataFrame([[0,'Not Distracted'],[1,'Looked But Did Not See'],[3,'By Other Occupants'],[4,'By a Moving Object In Vehicle'],[5,'While Talking Or Listening to Cellular Phone'],[6,'While Manipulating Cellular Phone'],[7,'While Adjusting Audio or Climate Controls'],[9,'While Adjusting Audio or Climate Controls'],[10,'While Using Or Recahing For Device'],[12,'Distracted By Outside Person'],[13,'Eating Or Drinking'],[14,'Smoking Related'],[15,'Other Cellular Phone Related'],[16,'No Driver Present'],[17,'Distracted'],[18,'Distracted'],[19,'Distracted'],[92,'Distracted'],[93,'Distracted'],[99,'Not Distracted']])
```

```python
grouped = big_raw.groupby(['CASENUM']).max()
grouped_viol = viol_raw.groupby(['MVIOLATN']).count()
grouped_viol
```

```python
grouped = grouped.replace({'MDRDSTRD':{0:'Not Distracted',96:'Not Distracted',99:'Not Distracted',17:'Distracted/Unknown Reason',18:'Distracted/Unknown Reason',19:'Distracted/Unknown Reason',92:'Distracted/Unknown Reason',93:'Distracted/Unknown Reason',98:'Distracted/Unknown Reason',1:'Looked But Did Not See',3:'By Other Occupants',4:'By a Moving Object In Vehicle',5:'Cell Phone',6:'Cell Phone',7:'Adjusting Audio or Climate Controls',9:'Adjusting Audio or Climate Controls',10:'Reaching ',12:'Distracted from outside',13:'Eating or Drinking',14:'Smoking Related',15:'Cell Phone',16:'No Driver',97:'Day Dreaming'}})
grouped = grouped.replace({'REGION':{1:'Northeast',2:'Midwest',3:'South',4:'West'}})
grouped = grouped.replace({'URBANICITY':{1:'Urban',2:'Rural'}})
```

```python
grouped_path = os.path.join('Results','distracted_clean.csv')
grouped.to_csv(grouped_path)
```

```python
quick_stats = grouped['MDRDSTRD'].value_counts()
quick_stats.to_frame()
quick_path = os.path.join('Results','distracted_stats.csv')
quick_stats.to_csv(quick_path)
```
